## SELECCIÓN
Una **estructura de selección** es una **declaración condicional**, el cual permite que diferentes secuencias de código se ejecuten dependiendo de si una **condición booleana** especificada se evalúa como **verdadera o falsa.**


### Ejemplo 1: Determinar calificación de un examen



```python
# Definir variable y condicional
entrada = input("Ingresa calificación (0-100) [por defecto 85]: ").strip()
calificacion = 85 if entrada == "" else int(entrada)

if calificacion < 0 or calificacion > 100:
    print("Valor fuera de rango. Debe estar entre 0 y 100.")
else:
    if calificacion >= 90:
        letra = 'A'
    elif calificacion >= 80:
        letra = 'B'
    elif calificacion >= 70:
        letra = 'C'
    elif calificacion >= 60:
        letra = 'D'
    else:
        letra = 'F'
    print(f"Calificación numérica: {calificacion} → Letra: {letra}")
```

### Ejemplo 2: Insertar números pares dentro de una lista



```python
# Definir lista y usar condicional para filtrar pares
numeros = list(range(1, 21))  # 1 a 20
pares = []
for n in numeros:
    if n % 2 == 0:  # condición par o no par
        pares.append(n)

print("Números originales:", numeros)
print("Números pares:", pares)
```

### Ejemplo 3: Arrojar la edad de una persona dentro de un diccionario



```python
# Definir variable y condicional
entrada = input("Ingresa tu edad (entero ≥ 0) [por defecto 30]: ").strip()
edad = 30 if entrada == "" else int(entrada)

persona = {}
if edad >= 0:
    persona["edad"] = edad
    print("Diccionario persona:", persona)
else:
    print("Edad no válida. Debe ser un entero mayor o igual a 0.")
```

## ITERACIÓN — Ciclo **for**



### Ejemplo 1: Sumar los números del 1 al 10


```python
# Realizar el ciclo for para sumar 1..10
suma = 0
for i in range(1, 11):
    suma += i
print("Suma de 1 a 10:", suma)
```

### Ejemplo 2: Insertar números secuenciales dentro de una lista


```python
# Definir variables y realizar ciclo for
inicio = 5
fin = 15
secuencial = []
for n in range(inicio, fin + 1):
    secuencial.append(n)
print(f"Lista con números de {inicio} a {fin}:", secuencial)
```

### Ejemplo 3: Imprimir las claves y valores de un diccionario


```python
# Definir el diccionario y recorrerlo
alumno = {"nombre": "Ana", "edad": 22, "carrera": "Ing. Sistemas"}
for clave, valor in alumno.items():
    print(f"{clave} → {valor}")
```

## ITERACIÓN — Ciclo **while**



### Ejemplo 1: Validar una cadena



```python
# Realizar el ciclo while de validación
nombre = ""
while not nombre.isalpha():
    nombre = input("Ingresa tu nombre (solo letras): ").strip()
    if nombre == "":
        print("El nombre no puede estar vacío.")
    elif not nombre.isalpha():
        print("El nombre solo debe contener letras (sin espacios ni números).")
print("Nombre válido:", nombre)
```

### Ejemplo 2: Recorrer las opciones de un menú



```python
# Definir variables y realizar el ciclo while
opcion = ""
while opcion != "4":
    print("""Menú principal:
1. Saludar
2. Mostrar hora
3. Ayuda
4. Salir""")
    opcion = input("Selecciona una opción (1-4): ").strip()
    if opcion == "1":
        print("¡Hola! 👋")
    elif opcion == "2":
        from datetime import datetime
        print("Hora actual:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    elif opcion == "3":
        print("Ingresa un número del 1 al 4 para ejecutar una acción.")
    elif opcion == "4":
        print("Saliendo del menú...")
    else:
        print("Opción no válida. Intenta de nuevo.")
```

### Ejemplo 3: Calcular la suma de los valores de un diccionario


```python
# Definir variables y usar while para recorrer por índice
productos = {"teclado": 350, "raton": 200, "monitor": 4200}
claves = list(productos.keys())
idx = 0
suma_valores = 0
while idx < len(claves):
    clave = claves[idx]
    suma_valores += productos[clave]
    idx += 1
print("Suma total de valores:", suma_valores)
```

## CHALLENGE: Agenda con operaciones CRUD
Una vez visto el ***Hands - On: Estructuras de Control de Flujo***, se presentan la siguientes actividades para que el alumnado pueda repasar y reforzar lo aprendido dentro de la clase.


Dentro de la siguiente dinámica, **se utilizará el challenge de la agenda** para **crear las operaciones de un CRUD** donde **dentro de los mismos se optimicen sus procesos haciendo uso de las estructuras de control de flujo**, todo lo anterior haciendo uso de **Python**.

Las operaciones esperadas del CRUD son:

*   **Ver agenda**
*   **Crear contacto**
*   **Consultar contacto**
*   **Actualizar contacto**
*   **Eliminar contacto**

### 1) Crear la agenda
Usaremos una **lista de diccionarios**, donde cada contacto tiene: `nombre`, `telefono` y `correo`.


```python
# Definir agenda (registros de ejemplo)
agenda = [
    {"nombre": "Chica", "telefono": "11-1111-1111", "correo": "chica@outlook.com"},
    {"nombre": "Cometa", "telefono": "22-2222-2222", "correo": "cometa@outlook.com"},
    {"nombre": "Margarito", "telefono": "33-3333-3333", "correo": "margarito@outlook.com"},
    {"nombre": "Buchona", "telefono": "44-4444-4444", "correo": "buchona@outlook.com"},
    {"nombre": "Nino", "telefono": "55-5555-5555", "correo": "nino@outlook.com"},
]
print("Agenda inicial con", len(agenda), "contactos.")
```

### 2) Imprimir todos los contactos de la agenda
Recorremos la lista con `for` y mostramos un **número de contacto** por iteración.


```python
# Mostrar todos los registros de la agenda
for i, c in enumerate(agenda, start=1):
    print(f"Contacto #{i}: nombre={c['nombre']}, teléfono={c['telefono']}, correo={c['correo']}")
```

### 3) Crear un contacto
Solicitamos datos por teclado, los guardamos en un diccionario y lo añadimos a la agenda.


```python
# Crear contacto
nombre = input("Nombre nuevo: ").strip()
telefono = input("Teléfono: ").strip()
correo = input("Correo: ").strip()

contacto = {"nombre": nombre, "telefono": telefono, "correo": correo}
agenda.append(contacto)
print("Contacto creado con éxito.")

# Mostrar el último contacto para confirmar
print("Creado:", agenda[-1])
```

### 4) Consultar un contacto
Buscamos por **nombre** con un `for`. Si lo encontramos, lo imprimimos; si no, avisamos.


```python
# Consultar contacto
buscado = input("Nombre a consultar: ").strip().lower()
encontrado = None
for c in agenda:
    if c["nombre"].lower() == buscado:
        encontrado = c
        break

if encontrado:
    print("Contacto encontrado:", encontrado)
else:
    print("Contacto no encontrado.")
```

### 5) Actualizar un contacto
Primero localizamos el contacto. Si existe, mostramos un **menú** con `while` para actualizar **nombre**, **teléfono** o **correo**.


```python
# Actualizar contacto
buscado = input("Nombre a actualizar: ").strip().lower()
indice = -1
for i, c in enumerate(agenda):
    if c["nombre"].lower() == buscado:
        indice = i
        break

if indice == -1:
    print("Contacto no encontrado.")
else:
    print("Contacto actual:", agenda[indice])
    while True:
        print("""Menú actualizar contacto:
1. Actualizar nombre
2. Actualizar teléfono
3. Actualizar correo
4. Salir del menú""")
        op = input("Elige una opción (1-4): ").strip()
        if op == "1":
            nuevo = input("Nuevo nombre: ").strip()
            if nuevo:
                agenda[indice]["nombre"] = nuevo
                print("Actualizado:", agenda[indice])
            else:
                print("El nombre no puede estar vacío.")
        elif op == "2":
            nuevo = input("Nuevo teléfono: ").strip()
            agenda[indice]["telefono"] = nuevo
            print("Actualizado:", agenda[indice])
        elif op == "3":
            nuevo = input("Nuevo correo: ").strip()
            agenda[indice]["correo"] = nuevo
            print("Actualizado:", agenda[indice])
        elif op == "4":
            print("Saliendo del menú de actualización...")
            break
        else:
            print("Opción no válida. Intenta de nuevo.")
```

### 6) Eliminar un contacto
Buscamos por nombre; si lo encontramos, lo **eliminamos** de la agenda.


```python
# Eliminar contacto
buscado = input("Nombre a eliminar: ").strip().lower()
indice = -1
for i, c in enumerate(agenda):
    if c["nombre"].lower() == buscado:
        indice = i
        break

if indice == -1:
    print("Contacto no encontrado.")
else:
    eliminado = agenda.pop(indice)
    print("Contacto eliminado:", eliminado)
    print("Total de contactos restantes:", len(agenda))
```
