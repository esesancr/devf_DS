## **FUNCIÓN INTEGRADA**
Una **función integrada** es un bloque de código reutilizable
 predefinida en python que se utiliza para **realizar una tarea específica y están disponibles para su uso en cualquier momento** sin la necesidad de importar ningún módulo.
Una función integrada tiene la siguiente estructura:
```
nombre_funcion(parametro1, parametro2, ...)
```

## **FUNCIÓN DEFINIDA POR EL USUARIO**
Una **función definida por el usuario** es creada de manera personalizada para realizar tareas específicas y reutilizables en todo el programa. Este tipo de función se crea utilizando la palabra clave ` def` seguida del nombre de la función y los parámetros, si los hay.
Una función definida por el usuario tiene la siguiente estructura:
```
def nombre_funcion(parametro1, parametro2, ...):
 # Código que realiza una tarea específica
 # Devolver un valor (opcional)
 return valor
```

### **EJEMPLO 1:**
Función que calcula el cuadrado de un número:


````python
# difinir función
num_cuad = int(input('Ingresa un número: '))
def cuadrado(numero):
 return numero ** 2
cuadrado(num_cuad)
````

### **EJEMPLO 2:**
Función para verificar si un número es par o impar:


````python
# difinir función
num_par = int(input('Ingresa un número: '))
def par(numero):
 if numero % 2 == 0:
  print('Par')
 else:
  print('Impar')
par(num_par)
````

### **EJEMPLO 3:**
Función para calcular el factorial de un número:


````python
# difinir función
num_fact = int(input('Ingresa un número: '))
def factorial(numero):
 if numero == 0:
  return 1
 else:
  return numero * factorial(numero - 1)
factorial(num_fact)
````

# **CHALLENGE: AGENDA CON MENÚ CRUD — Solución**


## 1) Definir agenda


````python
# Definir agenda
agenda = [
    {"Nombre": "Chica", "Telefono": "11-1111-1111", "Correo": "chica@outlook.com"},
    {"Nombre": "Cometa", "Telefono": "22-2222-2222", "Correo": "cometa@outlook.com"},
    {"Nombre": "Margarito", "Telefono": "33-3333-3333", "Correo": "margarito@outlook.com"},
    {"Nombre": "Buchona", "Telefono": "44-4444-4444", "Correo": "buchona@outlook.com"},
    {"Nombre": "Nino", "Telefono": "55-5555-5555", "Correo": "nino@outlook.com"},
]

````

## 2) Función de **menú CRUD**
Muestra las opciones disponibles.


````python
# Crear función menú opciones
def menu_opciones():
    print("Menú CRUD - Agenda:")
    print("1. Ver agenda")
    print("2. Crear contacto")
    print("3. Consultar contacto")
    print("4. Actualizar contacto")
    print("5. Eliminar contacto")
    print("6. Salir menú")
````

````python
# Crear función mostrar agenda
def mostrar_agenda(lista):
    if len(lista) == 0:
        print("Agenda vacía")
    else:
        for i, c in enumerate(lista, start=1):
            print(f"#{i} Nombre={c['Nombre']}, Teléfono={c['Telefono']}, Correo={c['Correo']}")
````

## 4) Función para **crear** contacto



````python
# Crear función crear contacto
def crear_contacto(lista):
    nombre = input("Nombre: ").strip()
    telefono = input("Teléfono: ").strip()
    correo = input("Correo: ").strip()
    contacto = {"Nombre": nombre, "Telefono": telefono, "Correo": correo}
    lista.append(contacto)
    print("Contacto creado con éxito")
````

## 5) Función para **consultar** contacto por nombre



````python
# Crear función consultar contacto
def consultar_contacto(lista):
    nombre = input("Nombre a consultar: ").strip().lower()
    encontrado = None
    for c in lista:
        if c["Nombre"].lower() == nombre:
            encontrado = c
            break
    if encontrado is not None:
        print("Contacto encontrado:", encontrado)
    else:
        print("Contacto no encontrado")
````

## 6) **Actualizar**


````python
# Crear función menú para actualizar contacto
def menu_actualizar():
    print("Menú actualizar contacto:")
    print("1. Actualizar nombre")
    print("2. Actualizar teléfono")
    print("3. Actualizar correo")
    print("4. Salir del menú")

# Crear función buscador contacto para actualizar contacto
def buscar_contacto(lista, nombre):
    for i, c in enumerate(lista):
        if c["Nombre"].lower() == nombre.lower():
            return i
    return None

# Crear función actualizar nombre para actualizar contacto
def actualizar_nombre(lista, indice):
    nuevo = input("Nuevo nombre: ").strip()
    if nuevo:
        lista[indice]["Nombre"] = nuevo
        print("Actualizado:", lista[indice])

# Crear función actualizar telefono para actualizar contacto
def actualizar_telefono(lista, indice):
    nuevo = input("Nuevo teléfono: ").strip()
    lista[indice]["Telefono"] = nuevo
    print("Actualizado:", lista[indice])

# Crear función actualizar correo para actualizar contacto
def actualizar_correo(lista, indice):
    nuevo = input("Nuevo correo: ").strip()
    lista[indice]["Correo"] = nuevo
    print("Actualizado:", lista[indice])
````

````python
# Crear función actualizar contacto
def actualizar_contacto(lista):
    nombre = input("Nombre a actualizar: ").strip()
    i = buscar_contacto(lista, nombre)
    if i is None:
        print("Contacto no encontrado")
    else:
        while True:
            menu_actualizar()
            op = input("Opción (1-4): ").strip()
            if op == "1":
                actualizar_nombre(lista, i)
            elif op == "2":
                actualizar_telefono(lista, i)
            elif op == "3":
                actualizar_correo(lista, i)
            elif op == "4":
                break
            else:
                print("Opción no encontrada")
````

## 7) Función para **eliminar** contacto



````python
# Crear función eliminar contacto
def eliminar_contacto(lista):
    nombre = input("Nombre a eliminar: ").strip()
    indice = None
    for i, c in enumerate(lista):
        if c["Nombre"].lower() == nombre.lower():
            indice = i
            break
    if indice is None:
        print("Contacto no encontrado")
    else:
        eliminado = lista.pop(indice)
        print("Contacto eliminado:", eliminado)
````

## 8) Función **menú agenda**



````python
# Crear función menú agenda
def menu_agenda():
    while True:
        menu_opciones()
        opcion = input("Opción (1-6): ").strip()
        if opcion == "1":
            mostrar_agenda(agenda)
        elif opcion == "2":
            crear_contacto(agenda)
        elif opcion == "3":
            consultar_contacto(agenda)
        elif opcion == "4":
            actualizar_contacto(agenda)
        elif opcion == "5":
            eliminar_contacto(agenda)
        elif opcion == "6":
            break
        else:
            print("Opción no encontrada")
````

````python
menu_agenda()
````
